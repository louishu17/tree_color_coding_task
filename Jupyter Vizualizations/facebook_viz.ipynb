```python
data = pd.read_csv('Similarity_Normalized.csv')
row = set()
data = data.loc[data['n'] == 7.0]
for x in data['File A']:
    row.add(x)
for x in data['File B']:
    row.add(x)
row = list(row)
matrix = np.zeros((98,98))
df = pd.DataFrame(matrix)
df.columns=row
df.index=row

for x in range(len(data)):
    df[data['File A'].iloc[x]][data['File B'].iloc[x]] = data['Similarity'].iloc[x]
    df[data['File B'].iloc[x]][data['File A'].iloc[x]] = data['Similarity'].iloc[x]
#sns.heatmap(df.iloc[68:84,41:57], cmap='Blues')
```

```python
#Basic K means clustering algorithm by Kieran
#Format dataset to score matrix

matrix = df
#Start clustering process
def univ_kmeans(K,matrix,size):
    schools = list(matrix.index)
    count = 0
    average_best = 0
    best_vals = [[0] for p in range(K)]
    best_clusters = [[] for p in range(K)]
    best_centers = 0
    for num in range(size):  
        x = (matrix.sample(n=K))
        x = list(x.index)
        clusters = [[] for p in range(K)]
        vals = [[] for p in range(K)]
        for school in matrix.index:
            if (school in x):
                position = x.index(school)
                clusters[position].append(school)
            else:
                scores = 0
                pos = 0
                for y in range(K):
                    new = matrix[school][x[y]]
                    if new > scores:
                        scores = new
                        pos = y
                vals[pos].append(scores)
                clusters[pos].append(school)
        average_current = 0
        for z in vals:
            if (len(z) > 0):
                average_current += sum(z)/len(z)
        average_current = average_current/K   
        if average_current > average_best:
            average_best = average_current
            best_clusters = clusters
            best_vals = vals
            best_centers = x
        count += 1
    for z in best_vals:
        print(sum(z)/len(z))
    print(clusters)
    print(best_centers)
    print(count)

univ_kmeans(3,matrix,1000)
```
