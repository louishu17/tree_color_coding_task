```python
#A file for final vizualizations
import nbformat
import pandas as pd 
import numpy as np 
import matplotlib as plot
from matplotlib import pyplot as plt
from nbconvert import HTMLExporter
from pandas import DataFrame
import seaborn as sns
```

```python
# A vizualization to show changing error for simulation 1
# t = 100, p = 0.8, tree = template.graph
sim1_n = []
sim1_err = []
sim1_out = []
sim1 = pd.DataFrame(list(zip(sim1_n,sim1_err,sim1_out)), columns = ['n','error','output'])
sim1['error'] = sim1['error'].abs()
sim1 = sim1[sim1['error'] < .15]
sns.scatterplot(data = sim1, x = 'n', y = 'error').set(xlabel = 'n', ylabel = 'Error', title = 'Error as N Changes')
sns.set_style("darkgrid", {'ytick.left' : True, 'xtick.bottom': True, 'axes.spines.left': True, 'grid.color': '1', 'axes.edgecolor': '0', 'axes.spines.right': False, "axes.facecolor": ".9"})
sns.set_context("paper")
sns.despine(right = True, top=True)
sns.set_palette("colorblind", 10)
sim1
```
